# Jeans scraper of my 10 pages data 

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
}
```

```python
brand_name=[]
price=[]
discount=[]
cloth_material=[]
fit_type=[]
length_type=[]
rating=[]
gender=[]

for page in range(1, 10):  
    print(f"Scraping page {page}...")
    url = f"https://www.amazon.in/s?k=jeans+for+men+and+women&i=apparel&page={page}"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    b = soup.find_all("span", class_="a-size-base-plus a-color-base")
    for i in b:
        brand_name.append(i.text.strip())
        
    p = soup.find_all("span", class_="a-price-whole")
    for i in p:
        price.append(i.text.strip())

    pattern = re.compile(r"\(\d+% ?off\)", re.IGNORECASE)
    s = soup.find_all("span", string=pattern)
    for i in s:
        discount.append(i.text.split("%")[0].split("(")[1])

    f = soup.find_all("h2", class_="a-size-base-plus a-spacing-none a-color-base a-text-normal")

    for i in f:
        text = i.text.strip()
        if re.search(r"\bWomen'?s?\b", text, re.I):
            gender.append("Women")
        elif re.search(r"\bMen'?s?\b", text, re.I):
            gender.append("Men")

        fit = re.search(r"\b(Slim|Regular|Relaxed|Baggy|Straight|Loose|Bootcut|Skinny|Tapered|Mom|Wide)\b", text, re.I)
        fit_type.append(fit.group(1).capitalize() if fit else None)

        ln = re.search(r"(Ankle Length|Full Length|Cropped|Regular Length)", text, re.I)
        length_type.append(ln.group(0) if ln else None)

        mat = re.search(r"(Cotton|Denim|Lycra|Stretchable|Blend|Fabric)", text, re.I)
        cloth_material.append(mat.group(0) if mat else None)

    r = soup.find_all("span", class_="a-size-small a-color-base")
    for i in r:
        text = i.text.strip()
        if "& Up" not in text:
            rating.append(text)

    time.sleep(2)

#to make all length equal
min_len = min(len(brand_name), len(price), len(discount), len(gender), len(fit_type), len(cloth_material), len(length_type), len(rating))
brand_name = brand_name[:min_len]
price = price[:min_len]
discount = discount[:min_len]
gender = gender[:min_len]
fit_type = fit_type[:min_len]
cloth_material = cloth_material[:min_len]
length_type = length_type[:min_len]
rating = rating[:min_len]

df = pd.DataFrame({
    "brand_name": brand_name,
    "price": price,
    "discount": discount,
    "gender": gender,
    "fit_type": fit_type,
    "cloth_material": cloth_material,
    "length_type": length_type,
    "rating": rating
})

print(df.shape)
print(df.head())
```
